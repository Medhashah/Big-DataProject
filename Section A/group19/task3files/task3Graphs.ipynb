```python
import csv
import glob
import numpy as np
import pandas as pd
from cycler import cycler
import matplotlib.cm as cm
import matplotlib.pyplot as plt
```

```python
with open('complaintsAllYears.csv','r') as f:
    content = f.readlines()
x=[]
y=[]
z=[]
df = []
for i in content:
    data = i.split(",")
    data[2] = int(data[2].strip("\n"))
    df.append(data)
df = pd.DataFrame(df,columns=['Borough','Complaint Type','No. of complaints'])
colors = plt.cm.jet(np.linspace(0, 1, 17))
df.pivot('Borough','Complaint Type','No. of complaints').plot(kind='bar',rot=0,color=colors)
plt.xlabel("Borough")
plt.ylabel("No. of complaints")
plt.legend(loc=(0.65,0.45), prop={'size': 4.5})
plt.tick_params(axis='x', which='major', labelsize=7)
#plt.figure(figsize=(200,200))
plt.savefig("allYears.jpg", dpi=300)
```

```python
content = []
for file in glob.glob('borough_avg_delay.csv/'+'*.csv'):
    with open(file,'r') as f:
        data = f.readlines()
        if data != []:
            content.append(data)
x=[]
y=[]
for i in content:
    data = i[0].split(",")
    x.append(data[0])
    y.append(float(data[1].strip("\n")))
plt.bar(x,y)
plt.tick_params(axis='x', which='major', labelsize=7)
plt.xlabel("Borough")
plt.ylabel("Average delay")
plt.savefig("borough_average_delay.jpg", dpi=300)
```
