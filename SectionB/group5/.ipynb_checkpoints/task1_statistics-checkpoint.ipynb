```python
import pandas as pd
import numpy as np
import json
import glob
import matplotlib.pyplot as plt
import os
from collections import Counter
```

```python
alldicts = []
f = "/Users/redviper/Documents/BigDataProject/task1_output/"
for file in os.listdir(f):
    full_filename = "%s%s" % (f, file)
    print(full_filename)
    with open(full_filename,'r') as fi:
        dict = json.load(fi)
        alldicts.append(dict)
```

```python
#define column datatypes: XXXX. From left to right: X is 1 if it contains type int, float, datetime, and text.
#For example: if column contains int and text, it will denote as 1001.
column_types = []
types = []
for d in alldicts:
    for c in d['columns']:
        code = 0
        data_types = c['data_types']
        for t in data_types:
            types.append(t['type'])
            if t['type'] == 'TEXT':
                code+=1
            elif t['type'] == 'DATE/TIME':
                code+=10
            elif t['type'] == 'REAL':
                code+=100
            elif t['type'] == 'INTEGER (LONG)':
                code+=1000
        column_types.append(code)
type_counts = Counter(types)
df = pd.DataFrame.from_dict(type_counts, orient='index')
df.plot.bar(legend=False, color='green')            
    
```
